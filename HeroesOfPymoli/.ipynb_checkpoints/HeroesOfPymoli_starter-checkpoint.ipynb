```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "purchase_data.csv"

# Read Purchasing File and store into Pandas data frame

purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
#Calculating the number of players using unique function to not double count a same user
total_num_pl = len(purchase_data["SN"].unique())
table_total = pd.DataFrame({'Total': [total_num_pl]})
table_total
```

## Purchasing Analysis (Total)

```python
item_total = len(purchase_data["Item ID"].unique())
avg_price = round(purchase_data["Price"].mean(),2)
purch_vote = purchase_data["Purchase ID"].count()
total_rev = purchase_data["Price"].sum()

summary_tb = pd.DataFrame({'Number of Unique Items': [item_total],
                            "Average Price" : [avg_price], 
                            "Purchase Total" : [purch_vote], 
                            "Total Revenue" : [total_rev]})
summary_tb
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#drop any duplicates in the SN column
sn_df = purchase_data.drop_duplicates('SN')

#count the values of each gender with in the new variable sn_df as duplicates were dropped
total_gender=sn_df['Gender'].value_counts()

#get total for each gender
male_total = total_gender['Male']
female_total = total_gender['Female']
other_total = total_gender['Other / Non-Disclosed']


#get total percent of each gender       
male_percent = total_gender['Male']/total_num_pl*100
female_percent = round(total_gender['Female']/total_num_pl*100, 2)
other_percent = total_gender['Other / Non-Disclosed']/total_num_pl*100


gender_summary_tb = pd.DataFrame({'Gender Demographics':['Male', 'Female', 'Other / Non-Disclosed'], 
                                  'Total Count':[male_total, female_total, other_total],
                                   'Percentage' : [male_percent, female_percent , other_percent]})

gender_summary_tb
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#purchase count by gender grouping by gender 
gender_list =purchase_data.groupby("Gender")

#the total gender count
total_gender = gender_list["Gender"].count()

#total gender count 
female_count = total_gender['Female']
male_count= total_gender['Male']
other_count = total_gender['Other / Non-Disclosed']

#purchase price average per gender
#gender_list_price =purchase_data.groupby("Price")
gender_price=gender_list['Price'].mean()

female_purchase= gender_price['Female']
male_purchase = gender_price['Male']
other_purchase = gender_price ['Other / Non-Disclosed']

#total purchase sum 
gender_price=gender_list['Price'].sum()
female_sum = gender_price['Female']
male_sum = gender_price['Male']
other_sum = gender_price ['Other / Non-Disclosed']

#Avg Total Purchase_per person
male_avg_p = male_sum/male_total
female_avg_p = female_sum/female_total
other_avg_p = other_sum/other_total

purchasing_summary_tb = pd.DataFrame({'Gender':['Male', 'Female', 'Other / Non-Disclosed'], 
                                  'Purchase Count':[male_count, female_count, other_count],
                                   'Average Purchase Price': [male_purchase, female_purchase , other_purchase],
                                    'Total Purchase Value': [male_sum, female_sum , other_sum],
                                     'Avg Total Purchase per Person': [male_avg_p, female_avg_p, other_avg_p]})

purchasing_summary_tb 
```
